```python
import pandas as pd
import numpy as np
import datetime as dt
```

**Read Datasets**

```python
df_movies = pd.read_csv("movies_metadata.csv") #from Kaggle
df_ratings = pd.read_csv("ratings_small.csv") #subset from Kaggle website for ratings
```

```python
df_ratings.head() #make sure they output properly
```

```python
df_movies.head()
```

```python
df_movies.columns,df_movies.dtypes #want to transform movie release_date from object to date_time
```

```python
#Clean the data
df_ratings = df_ratings.drop("timestamp",axis=1) #don't care about the timestamp, it not associated with a list
df_ratings["rating"] = df_ratings["rating"].astype("int") #rounding up the ratings
df_movies = df_movies[df_movies["id"].str.isnumeric()] #compare the movie_ids from objects to integers (some are letters too by mistake) 
df_movies["id"] = df_movies["id"].astype("int") #convert the movie_ids to integer

#QUESTIONG: How should we populate missing dates???
df_movies.release_date = df_movies.release_date.fillna('1989-12-31') #if the release_date empty make it 12-1989
df_movies = df_movies.drop(df_movies[pd.to_datetime(df_movies.release_date,errors="coerce").isna()].index) #drop any item it cannot convert to date_time
#anything that cannot be converted --> convert to NA --> return all the NA values.
df_movies.release_date = pd.to_datetime(df_movies.release_date) #finishing the conversion after the drop of all the values remaining
```

```python
#TEST: How many rows I couldn't convert to date_time for release_date (rows = 90)
df_movies[pd.to_datetime(df_movies.release_date,errors="coerce").isna()].shape
```

```python
#Merge the two datasets using an inner to prevent orphaned ids??
#Merge based on the id from MOVIES & movie_id from RATINGS. 
#Inner: I get rid of ids don't match in both tables. Only keep the intersection.
#The intersection is in movies. df_ratings table stays the same.
df_movies = df_movies.merge(df_ratings,left_on="id",right_on="movieId",how="inner")
```

```python
df_movies.head()
```

```python
df_movies_lite = df_movies[["userId","movieId","rating","vote_count", "release_date", "runtime", "revenue"]]
df_movies_lite
```

```python
#Create the ratings.csv for librec-auto
df_ratings = df_movies[["userId","movieId","rating"]] #overwrite the existing df_ratings
df_ratings.to_csv("ratings.csv",index=None,header=False)
```

```python
#creating the protected feature bins
avg_vote = df_movies.vote_count.mean() #popular vs unpopular (higher than avge is popular, otherwise unpopular)
avg_rev = df_movies.revenue.mean() #higher_revenue vs low_revenue
avg_runtime = df_movies.runtime.mean() #longer vs shorter
cutoff_date = dt.datetime.strptime("1990","%Y") #cutoff date is 1990 old vs new
```

```python
#create dummies for each protected feature
#.transform for each value transform this line. Lambda is a simple instant function
#The function inputs x. One by one applies the vote_count to the function.
#get_dummies panda function that converts categorical values into column names and it gives boolean (vectorizes each column)
df_pop = pd.get_dummies(df_movies.vote_count.transform( lambda x: "popular" if x >= avg_vote else "unpopular"))
df_date =  pd.get_dummies(df_movies.release_date.transform( lambda x: "old" if x >= cutoff_date else "new"))
df_rev =  pd.get_dummies(df_movies.revenue.transform( lambda x: "higher_revenue" if x >= avg_rev else "lower_revenue"))
df_len =  pd.get_dummies(df_movies.runtime.transform( lambda x: "longer" if x >= avg_runtime else "shorter"))
```

```python
#create full item feature table
#concatenating the protected features dummies into one dataset
df_item_features = pd.concat([df_movies["id"],df_pop,df_date,df_rev,df_len], axis=1) 
#No need for the ratings, so I can drop duplicates
df_item_features = df_item_features.drop_duplicates()
#stack makes the columns, places in the same column ("feature")
df_item_features = df_item_features.set_index("id").stack().reset_index()
df_item_features.columns = ["item_id","feature","value"]
df_item_features.head()
```

```python
#Removes all the features that are 0 value.
mask = df_item_features["value"]==1
df_item_features[mask]
```

```python
#Single out specifc feature
#another mask: for new or old feature
df_item_features = df_item_features[(df_item_features["feature"]=='old') | (df_item_features["feature"]=='new') ]
```
